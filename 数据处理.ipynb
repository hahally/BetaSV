```python
import pandas as pd
```

```python
train = pd.read_csv('../data/train.csv')
train
```

```python
train_mic = train[train.pid==7].reset_index(drop=True)
train_mic
```

```python
train_1m = train[(train.pid==2)&(train.mid==2)].reset_index(drop=True)
train_1m
```

```python
def read_file(file):
    lines = []
    with open(file=file,mode='r',encoding='utf-8') as f:
        line = f.readline().strip()
        while line:
            lines.append(line)
            line = f.readline().strip()
    
    return lines

def save_file(file, lines):
    with open(file=file,mode='w',encoding='utf-8') as f:
        for line in lines:
            f.write(line+'\n')
```

```python
train_1m_mic = pd.concat([train_1m,train_mic],axis=0).reset_index(drop=True)
train_1m_mic.to_csv('train_1m_mic.csv', index=False)
```

```python
train_all_spk2utt = [[spk]+train_1m_mic[train_1m_mic.spk==spk].utt.tolist() for spk in train_1m_mic.spk.unique()]
train_all_spk2utt = [' '.join(s) for s in train_all_spk2utt]
train_all_utt2spk = (train_1m_mic.utt + [' '] + train_1m_mic.spk).tolist()
train_all_scp = (train_1m_mic.utt + [' '] + train_1m_mic.filename).tolist()
```

```python
save_file(file='train_all_spk2utt',lines=train_all_spk2utt)
save_file(file='train_all_utt2spk',lines=train_all_utt2spk)
save_file(file='train.all.scp',lines=train_all_scp)
```

```python
trails_dev_1m = read_file('../data/trails_dev_1m')
lines = []
for line in trails_dev_1m:
    key = int(line.split()[0].split('_')[0] == line.split()[1].split('_')[0])
    lines.append(f'{key} {line}')
save_file(file='../data/trails_dev_1m',lines=lines)
```
